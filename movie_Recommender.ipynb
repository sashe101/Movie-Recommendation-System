```python
# ['Action','Adventure','Fantasy','Science Fiction']
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): # to convert string of list to list
       L.append(i['name'])
    return L
```

```python
# for cast col, extract top 3 dictionaries and from that only the actors name
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj): # to convert string of list to list
        if counter != 3:
           L.append(i['name'])
           counter+=1
        else:
            break
    return L
```

```python
# better approach is convert tags into vectors i.e text-->vectors process is known as text vectorization
# recommend closest vectors
# technique for text vectorization --> 'Bag of words'(using here}

# Bag of words :
# concatenate all tags i.e largeText= tag1+tag2+..............
# among the words in largeText find the top 5k most occurring words 
#            w1,w2,w3........w5000
# fr movie1: 5, 4 ,2,.......... ( no of times the words(w1,w2...) are occuring in movie1)
# fr movie2: 3, 5, 1...........
#....
#fr movie5k:....................
# table of (5000,5000) where each row here is known as vector
# if only 2 words then table of (5000,2) is in 2 dimensional space:
#            w1,w2
# fr movie1: 5, 4 (plot 5,4 in 2-dimensional space)
# fr movie2: 3, 5 (plot 3,5 in 2-dimensional space)
#....
#fr movie5k:.....

# we won't consider stopwords (eg->are,and,from,etc ; stopwords are used only for sentence formation) 
# so first remove stopwords & do vectorization on the rest 
# for this we ll use scikit learn --> inside scikit learn we have count vectorization that we ll use for this purpose
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# in higher dimension euclidean dist is not a reliable measure so here we use cosine distance
# cosine dist--> angle btw the vectors , the more the diff btw those angles the more distant those movies are
from sklearn.metrics.pairwise import cosine_similarity # cosine similarity--> always btw 0 & 1 
```

```python
similarity=cosine_similarity(vectors)
```

```python
 # if a movie is given then return 5 similar movies lyk the given movie
def recommend(movie):
    try:
        # Find the index of the movie in the DataFrame
        movie_index = new_df[new_df['title'] == movie].index[0]
        
        # Get the pairwise similarity scores for this movie
        distances = similarity[movie_index]
        
        # Get the indices of the top 5 similar movies (excluding the movie itself)
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        
        # Return the titles of the similar movies
        return [new_df.iloc[i[0]].title for i in movies_list]
    
    except IndexError:
        return f"Movie '{movie}' not found."
    except Exception as e:
        return str(e)
```

```python
recommend('Avatar')
```

```python
import gradio as gr

# Gradio function wrapper
def gradio_recommend(movie):
    # Get the list of similar movies
    similar_movies = recommend(movie)
    
    # Join the list into a single string with new lines
    if isinstance(similar_movies, list):
        return "\n".join(similar_movies)
    else:
        return similar_movies  # In case of error message

# Define Gradio inputs and outputs
inputs = gr.Textbox(label="Movie Title")
outputs = gr.Textbox(label="Recommended Movies")

# Create Gradio interface
interface = gr.Interface(
    fn=gradio_recommend,
    inputs=inputs,
    outputs=outputs,
    title="Movie Recommendation System",
    description="Enter a movie title and get the top 5 similar movies."
)

# Launch Gradio interface
interface.launch()
```
